## 9.1 matplotlib API入门

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
data=np.arange(10)
plt.plot(data)
```

### 9.1.1 Figure和Subplot

```python
fig=plt.figure()

ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(), 'k--')
ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30)+3*np.random.randn(30))
```

```python
fig, axes=plt.subplots(2,3)
axes
```

### 9.1.2 调整subplot周围的间距

```python
fig, axes=plt.subplots(2,2, sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k',alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```

### 9.1.3颜色,标记和线型

```python
plt.plot(np.random.randn(30).cumsum(), 'ko--')
```

```python
data=np.random.randn(30).cumsum()
plt.plot(data, '--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend()
```

### 9.1.4 刻度,标签和图例

```python
fig, axes=plt.subplots(2,1)
axes[0].plot(range(3))
print(plt.xlim())
print(plt.xlim([0,10]))
print(axes[0].get_xlim())
print(axes[0].set_xlim([-1,3]))
```

### 9.1.5 设置标题,轴标签,刻度以及刻度标签`

```python
plt.plot(np.random.randn(1000).cumsum())
plt.xticks([0,250, 500, 750, 1000],
           labels=['one', 'two', 'three', 'four', 'five'],
           rotation=30, fontsize='small')
plt.xlabel( 'Stages')
plt.title('My first matplotlib plot')
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())
ax.set_xticks([0,250,500, 750, 1000])
ax.set_xticklabels(['one','two','three','four','five'], rotation=30,fontsize='small')
ax.set_title('My first matplotlib plot')
ax.set_xlabel('Stages')
```

### 9.1.6 添加图例

```python
plt.plot(np.random.randn(1000).cumsum(), 'k', label='one')
plt.plot(np.random.randn(1000).cumsum(), 'k--', label='two')
plt.plot(np.random.randn(1000).cumsum(), 'k.', label='three')
plt.legend(labels=['one', 'two'])
```

### 9.1.7注解以及在subplot上绘图

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.text(0.5, 0.5, 'Hello World', family='monospace', fontsize=10)
```

```python
from datetime import datetime
import pandas as pd
data=pd.read_csv(r"D:\Github\pydata-book\examples\spx.csv",index_col=0, 
                 parse_dates=True)
spx=data.SPX
spx.plot(style='k-')
crisis_data=[(datetime(2007, 10, 11), 'Peak of bull market'),
             (datetime(2008, 3, 12), 'Bear Stearns Fails'),
             (datetime(2008, 9, 15), 'Lehman Bankruptcy')]
for date, label in crisis_data:
    plt.annotate(label, xy=(date, spx.asof(date)+75), 
                xytext=(date, spx.asof(date)+225),
#                  arrowprops=dict(arrowstyle='simple'),
                arrowprops=dict(facecolor='black', headwidth=4,
                                 width=2,headlength=4),
                horizontalalignment='left', verticalalignment='top')
plt.xlim(['1/1/2007', '1/1/2011'])
plt.ylim([600, 1800])
```

```python
fig, ax=plt.subplots(1,1)

rect=plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.2)
circ=plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon=plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color='g', alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)

plt.savefig('Shape_Drawing.png')
fig.savefig('Shape_Tight.png',dpi=400, bbox_inches='tight')
```

## 9.2 使用pandas和seaborn绘图

### 9.2.1 线型图

```python
s=pd.Series(np.random.rand(10).cumsum(), index=range(0, 100, 10))
s.plot()
```

```python
df=pd.DataFrame(np.random.randn(10,4).cumsum(0),
                columns=['A', 'B', 'C', 'D'],
                index=range(0, 100, 10))
df.plot()
```

### 9.2.2 柱状图

```python
fig, axes=plt.subplots(2, 1)
data=pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))

data.plot.bar(ax=axes[0], color='k', alpha=0.7)
data.plot.barh(ax=axes[1], color='k', alpha=0.7)
```

```python
#堆积柱状图表示展示百分比
tips=pd.read_csv(r"D:\Github\pydata-book\examples\tips.csv")
tips.head()
```

```python
party_counts=pd.crosstab(tips['day'], tips['size'])
party_counts
```

```python
party_counts=party_counts.loc[:, 2:5]
party_pcts=party_counts.div(party_counts.sum(1), axis=0)
party_pcts
```

```python
party_pcts.plot.bar(stacked=True)
```

```python
import seaborn as sns
sns.set(style='whitegrid')
tips['tip_pct']=tips.tip/(tips.total_bill - tips.tip)
sns.barplot(x='tip_pct', y='day', data=tips, orient='h')
```

```python
sns.barplot(x='tip_pct', y='day', hue='time',
            data=tips, orient='h', 
            order=['Thur', 'Fri', 'Sat', 'Sun'])
```

### 9.2.3 直方图和密度图

### 9.2.4 散布图或点图

```python
macro=pd.read_csv(r"D:\Github\pydata-book\examples\macrodata.csv")
data=macro[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data=np.log(data).diff().dropna()
trans_data[-5:]
```

```python
sns.regplot('m1', 'unemp', data=trans_data)
plt.title('Changes in log %s versus log %s' % ('m1','unemp'))
```

```python
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha':0.2})
```

### 9.2.4 分面风格和类型数据

```python
sns.factorplot(x='day', y='tip_pct', hue='time', 
               col='smoker', kind='bar',
               data=tips[tips.tip_pct<1])
```

```python
sns.catplot(x='day', y='tip_pct', row='time',
            col='smoker', kind='bar',
            data=tips[tips.tip_pct<1])
```

```python
sns.catplot(x='day', y='tip_pct', kind='box',
            data=tips[tips.tip_pct<0.5])
```

```python
sns.catplot(y='day', x='tip_pct', kind='box',
            data=tips[tips.tip_pct<0.5])
```
