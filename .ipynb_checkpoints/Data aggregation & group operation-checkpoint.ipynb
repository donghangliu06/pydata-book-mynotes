# 10 Data aggregation & group operation

## 10.1 Groupby机制

```python
import pandas as pd
df=pd.DataFrame({'key1': ['a','a','b','b','a'],
                 'key2': ['one','two','one','two','one'],
                 'data1': np.random.randn(5),
                 'data2': np.random.randn(5)})
df
```

```python
grouped=df['data1'].groupby(df['key1'])
grouped.mean()
```

```python
means=df['data1'].groupby([df['key1'], df['key2']]).mean()
means
```

```python
means.unstack()
```

```python
states=['Ohio','California','California','Ohio','Ohio']
years=[2005,2005,2006,2005,2006]
df['data1'].groupby([states, years]).mean()
```

```python
df.groupby('key1').mean()
```

```python
df.groupby(['key1','key2']).mean()
```

```python
df.groupby(['key1', 'key2']).size()
```

### 10.1.1 对分组进行迭代

```python
for name, group in df.groupby('key1'):
    print(name)
    print(group)
```

```python
for (k1,k2), group in df.groupby(['key1','key2']):
    print((k1,k2))
    print(group)
```

```python
pieces=dict(list(df.groupby('key1')))
pieces['b']
```

```python
list(df.groupby('key1'))
```

```python
df.dtypes
```

```python
grouped=df.groupby(df.dtypes, axis=1)
for dtype, group in grouped:
    print(dtype)
    print(group)
```

### 10.1.2 选取一列或列的子集

```python
df.groupby(['key1','key2'])[['data2']].mean()
```

```python
s_grouped=df.groupby(['key1','key2'])['data2']
s_grouped.mean()
```

```python
for (k1, k2), group in s_grouped:
    print((k1,k2))
    print(group)
```

### 10.1.3 通过字典或Series进行分组

```python
import numpy as np
people=pd.DataFrame(np.random.randn(5,5),
                    columns=list('abcde'),
                    index=['Joe','Steve','Wes','Jim','Travis'])
people.iloc[2:3, [1,2]]=np.nan
people
```

```python
mapping={'a':'red', 'b':'red', 'c':'blue',
         'd':'blue', 'e':'red', 'f':'orange'}
by_column=people.groupby(mapping, axis=1)
by_column.sum()
```

```python
map_series=pd.Series(mapping)
map_series
```

```python
people.groupby(map_series, axis=1).count()
```

### 10.1.4 通过函数进行分组

```python
for length, group in people.groupby(len):
    print(length)
    print(group)
```

```python
people.groupby(len).sum()
```

```python
key_list=['one']*3+['two']*2
people.groupby([len, key_list]).min()
```

```python
for (k1,k2),group in people.groupby([len, key_list]):
    print((k1,k2))
    print(group)
```

### 10.1.5 根据索引级别分组

```python
columns=pd.MultiIndex.from_arrays([['US']*3+['JP']*2,
                                   [1,3,5,1,3]],
                                  names=['cty','tenor'])
hier_df=pd.DataFrame(np.random.randn(4,5), columns=columns)
hier_df
```

```python
hier_df.groupby(level='cty', axis=1).count()
```

```python
hier_df.groupby(level='cty', axis=1).size()
```

### 10.2.1 面向列的多函数应用

```python
tips=pd.read_csv(r"D:\Github\pydata-book\examples\tips.csv")
tips['tip_pct']=tips['tip']/tips['total_bill']
tips[-5:]
```

```python
grouped=tips.groupby(['day', 'smoker'])
grouped_pct=grouped['tip_pct']
grouped_pct.agg('mean')
```

```python
grouped_pct.agg(['mean', 'std', lambda x:x.max()-x.min()])
```

```python
grouped_pct.agg([('foo','mean'), ('bar', 'std')])
```

```python
grouped_pct.agg({'foo':'mean'})
```

```python
functions=['count', 'mean', 'max']
result=grouped['tip_pct', 'total_bill'].agg(functions)
result
```

```python
grouped['tip_pct'].agg(['count','size'])
```

```python
ftuples=[('Durchschnitt','mean'), ('Abweichung', 'var')]
grouped['tip_pct', 'total_bill'].agg(ftuples)
```

```python
grouped.agg({'tip_pct':['min','max','mean','std'],
             'size':'sum'})
```

### 10.2.2 以"没有索引"的形式返回聚合数据

```python
tips.groupby(['day','smoker'], as_index=False).mean()
```

## 10.3 apply:一般性的拆分-应用-合并

```python
def top(df, n=5, column='tip_pct'):
    return df.sort_values(by=column)[-n:]
top(tips, n=6)
```

```python
tips.groupby('smoker').apply(top)
```

```python
tips.groupby(['smoker','day']).apply(top,n=1,column='total_bill')
```

```python
result=tips.groupby('smoker')['tip_pct'].describe()
```
