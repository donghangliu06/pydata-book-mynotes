## 8.2 合并数据集

### 8.2.1数据库风格的DataFrame合并 

```python
df1=pd.DataFrame({'key':['b', 'b', 'a', 'c', 'a', 'a','b'],
                  'data1':range(7)})
df2=pd.DataFrame({'key':['a', 'b', 'd'],
                  'data2':range(3)})
df1
```

```python
df2
```

```python
#merge默认将重叠列的列名当作联结键,内联结
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how='outer')
```

```python
left=pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                   'key2': ['one', 'two', 'one'],
                   'lval': [1, 2, 3]})
right=pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                    'key2': ['one', 'one', 'one', 'two'],
                    'rval': [4,5,6,7]})
pd.merge(left, right, on=['key1', 'key2'], how='outer')
```

```python
pd.merge(left, right, on='key1')
```

```python
pd.merge(left, right, on='key1', suffixes=('_left', '_right'))
```

### 8.2.2 索引上的合并

```python
left1=pd.DataFrame({'key': ['a', 'b','a','a','b','c'],
                    'value': range(6)})
right1=pd.DataFrame({'group_val': [3.5, 7]},
                    index=['a', 'b'])
left1
```

```python
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
lefth=pd.DataFrame({'key1': ['Ohio', 'Ohio', 'Ohio','Nevada','Nevada'],
                    'key2': [2000, 2001, 2002, 2001,2002],
                    'data': np.arange(5.)})
righth=pd.DataFrame(np.arange(12).reshape((6,2)),
                    index=[['Nevada', 'Nevada','Ohio','Ohio', 'Ohio','Ohio'],
                           [2001, 2000, 2000, 2000, 2001, 2002]],
                    columns=['event1', 'event2'])
lefth
```

```python
righth
```

```python
pd.merge(lefth, righth, left_on=['key1', 'key2'], right_index=True)
```

```python
pd.merge(lefth, righth, left_on=['key1', 'key2'], right_index=True, how='outer')
```

```python
left1
```

```python
right1
```

```python
left1.join(right1, on='key', how='outer') #以right1的索引,left1的on联结(默认索引)
```

```python
left1.join(right1, on='key') #默认左联结
```

```python
left1.join(right1, on='key', how='right')
```

```python
left2=pd.DataFrame([[1, 2],
                    [3, 4],
                    [5, 6]], index=['a', 'c', 'e'], columns=['Ohio', 'Nevada'])
right2=pd.DataFrame([[7, 8],
                     [9, 10],
                     [11, 12],
                     [13, 14]], index=['b', 'c', 'd', 'e'], columns=['Missouri','Alabama'])
another=pd.DataFrame([[7, 8],
                      [9, 10],
                      [11, 12],
                      [16, 17]], index=['a', 'c', 'e', 'f'], columns=['New York', 'Oregon'])
left2.join([right2, another])
```

```python
left2.join([right2, another], how='outer', sort=False)
```

### 8.3.1 重塑层次化索引

```python
data=pd.DataFrame(np.arange(6).reshape(2, 3),
                  index=pd.Index(['Ohio', 'Colorado'], name='state'),
                  columns= pd.Index(['one', 'two', 'three'], name='number'))
data
```

```python
result=data.stack()
result
```

```python
result.unstack()
```

```python
result.unstack(0)
```

```python
result.unstack('state')
```

```python
s1=pd.Series([0,1,2,3], index=['a', 'b', 'c', 'd'])
s2=pd.Series([4,5,6], index=['c', 'd', 'e'])
data2=pd.concat([s1, s2], keys=['one', 'two'])
data2
```

```python
data2.unstack()
```

```python
#stack会滤除缺失值
data2.unstack().stack()
```

```python
data2.unstack().stack(dropna=False)
```

### 8.3.2 将长格式旋转为宽格式

```python
data=pd.read_csv(r"D:\Github\pydata-book\examples\macrodata.csv")
data.head()
```

```python
periods=pd.PeriodIndex(year=data.year, quarter=data.quarter, name='date')
periods
```

```python
columns=pd.Index(['realgdp', 'infl', 'unemp'], name='item')
data=data.reindex(columns=columns)
```

```python
data.index=periods.asfreq('D', 'end')
data.head()
```

```python
ldata=data.stack().reset_index().rename(columns={0: 'value'})
ldata.head()
```

```python
ldata.set_index(['date', 'item']).unstack('item').value.head()
```

```python
ldata.pivot('date', 'item', 'value').head()
```

### 8.3.3 将宽格式旋转为长格式

```python
df=pd.DataFrame({'key': ['foo', 'bar', 'baz'],
                 'A': [1,2,3],
                 'B': [4,5,6],
                 'C': [7,8,9]})
df
```

```python
df.set_index('key').stack().reset_index().rename(columns={'level_1': 'variable',
                                                       0: 'value'})
```

```python
df.melt(['key'])
```

```python
df.melt(['key'], ['A', 'B'])
```

```python
df.melt(['key']).pivot('key', 'variable', 'value')
```

```python
df.melt(['key']).pivot('key', 'variable', 'value').reset_index()
```
